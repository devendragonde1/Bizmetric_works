```python
class Student_enquiry:
    Core_Fee=200000
    Students={}
    def __init__(self):
        self.course=['HR', 'Finance', 'Marketing', 'DS']
        self.analytics_fee = 0
        self.hostel_fee = 0
        self.food_fee = 0
        self.transport_fee = 0
        self.Total_fees=Student_enquiry.Core_Fee
    def user_input(self):
        self.user_name=input("Enter your name: ").capitalize()
        while True:
              self.sub=input("Enter the course you want: ").upper()
              if self.sub in self.course:
                   break
              else:
                   print("invalid course, Try again: ")         
        self.Analytics_pref=input("Enter your choice  for Analytics Y for Yes and N for No: ").upper()
        self.Hostel_opt=input("Enter choice for hostel Y for Yes and N for No: ").upper()
        self.Food_duration=int(input("Enter the duration for Hostel food :"))
        self.Transportation=input("Enter Transportation type Semester or Annual: ").upper()
        Student_enquiry.Students[self.user_name]={
             "Subject":self.sub,
             "Analytics":self.Analytics_pref,
             "Hostel":self.Hostel_opt,
             "FoodDuration":self.Food_duration,
             "Transportation_type":self.Transportation
        }
        return Student_enquiry.Students
    
```

```python
class Fee_calc(Student_enquiry):
    def __init__(self):
        super().__init__()
    def Analytic_fee(self):
        if self.Analytics_pref=='Y' and self.sub!='DS':
            self.analytics_fee=Student_enquiry.Core_Fee*0.10
            self.Total_fees+=self.analytics_fee
        else:
            print("No analytics for DS")
    def Hostel_fee(self):
        if self.Hostel_opt=='Y':
            self.hostel_fee=200000
            self.Total_fees+=self.hostel_fee
        else:
            print("You chose not to get a hostel.")
    def Food_fee(self):
        if self.Food_duration!=0:
            self.food_fee=self.Food_duration*2000
            self.Total_fees+=self.food_fee
    def Transportation_fee(self):
        if self.Transportation=='SEMESTER':
            self.transport_fee=13000
        elif self.Transportation=='ANNUAL':
            self.transport_fee=13000*2
        self.Total_fees+=self.transport_fee
    def Show_final_fees(self):
        print("\n------ FEE STRUCTURE ------")
        print("Name:", self.user_name)
        print("Course:", self.sub)
        print("Core Fee:", Student_enquiry.Core_Fee)
        print("Analytics Fee:", self.analytics_fee)
        print("Hostel Fee:", self.hostel_fee)
        print("Food Fee:", self.food_fee)
        print("Transport Fee:", self.transport_fee)
        print("----------------------------")
        print("Total Annual Cost:", self.Total_fees)
```

```python
Stu_1=Fee_calc()
Stu_1.user_input()
Stu_1.Analytic_fee()
Stu_1.Hostel_fee()
Stu_1.Food_fee()
Stu_1.Transportation_fee()
Stu_1.Show_final_fees()
```
