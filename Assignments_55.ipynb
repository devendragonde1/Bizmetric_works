```python
'''
Q 11.	Ask user to enter the marks and define if the user got distinction, first class, second class, pass, Fails
If marks are more than 80 output must be “You got distinction”
For marks more than 60 output must be “You got First class”
Marks more than 40 will display “You got second class”
Marks more than or equal to 35 “PASS” Otherwise Fail
'''

def remark(user_marks):
    if (not user_marks.isalpha and not user_marks.isalnum ) or user_marks.isdigit():
        markss=float(user_marks)
        if markss>100 or markss<0:
            return "enter marks in valid range"
        elif markss>=80:
            return "distinction"
        elif markss>=60:
            return "first class"
        elif markss>=40:
            return "Second class"
        elif markss>=35:
            return"pass"
        else:
            return "fail"
    else:
        return "enter marks only"
    
# main
marks=input("enter your marks :")
remark(marks)
```

```python
# Q14 using functions
Books={"HINDI":{"1st-4th":60,"5th-8th":100,"9-10th":150},
       "MARATHI":{"1st-4th":60,"5th-8th":100,"9-10th":150},
       "ENGLISH":{"1st-4th":80,"5th-8th":100,"9-10th":150},
       "SCIENCE":{"1st-4th":90,"5th-8th":120,"9-10th":200},
       "MATHS":{"1st-4th":100,"5th-8th":140,"9-10th":250}
       }

Notebooks={"SQUARE":{"100PGS":40,"200PGS":70},
            "4LINES":{"100PGS":30,"200PGS":50},
            "2LINES":{"100PGS":30,"200PGS":50},
            "SINGLELINE":{"100PGS":60,"200PGS":100},
            "A4 NOTEBOOK":{"100PGS":100,"200PGS":180}
            }

def user_query():
   order={}
   ch=int(input(  
                  """ Enter your choice amongs these 
                  1. books
                  2. notebooks
                  3. books and notebooks  enter number (1 or 2 or 3)"""
                  ))
   if ch in (1,3):
      order["books"] = {
            "subject": input("Enter subject: ").strip().upper(),
            "standard": int(input("Enter standard (1-10): ")),
            "quantity": int(input("Enter book quantity: "))
        }
   if ch in (2,3):
      pages = int(input("Enter pages (100 or 200): "))
      order["notebooks"] ={
            "type": input("Enter notebook type: ").strip().upper(),
            "pages": str(pages) + "PGS",
            "quantity": int(input("Enter notebook quantity: "))
        }
      
   return order

def user_bill(order):
    bill = {'details': {}, 'Total_amount': 0}

    #for books
    if "books" in order:
        std = order['books']['standard']

        if 1 <= std <= 4:
            std_key = "1st-4th"
        elif 5 <= std <= 8:
            std_key = "5th-8th"
        elif 9 <= std <= 10:
            std_key = "9-10th"
        else:
            print("Invalid standard")
            return

        subject = order['books']['subject']

        if subject not in Books:
            print("Invalid subject")
            return

        price = Books[subject][std_key] * order["books"]["quantity"]
        bill["details"]["books"] = price
        bill["Total_amount"] += price

    # ---------- NOTEBOOKS ----------
    if "notebooks" in order:
        n_type = order["notebooks"]["type"]
        pages = order["notebooks"]["pages"]

        if n_type not in Notebooks:
            print("Invalid notebook type")
            return

        price = (
            Notebooks[n_type][pages]
            * order["notebooks"]["quantity"]
        )
        bill["details"]["notebooks"] = price
        bill["Total_amount"] += price

    return bill

def print_bill(bill):
    for item, amount in bill["details"].items():
        print(f"{item.capitalize()} Amount: {amount}")
    print("Total Amount: ₹", bill["Total_amount"])

order=user_query()
bills=user_bill(order)
if input("do you want the bill Yes/No?").lower()=='yes':
    print_bill(bills)
```

```python
def salary_update(curr_salary, curr_grade):

    if curr_salary.replace('.', '', 1).isdigit():  
        curr_salary = float(curr_salary)

        if curr_salary <= 5:
            if curr_grade == 'A':
                return curr_salary * 1.16
            elif curr_grade == 'B':
                return curr_salary * 1.12
            elif curr_grade == 'C':
                return curr_salary * 1.10
            elif curr_grade == 'D':
                return curr_salary * 1.06
            else:
                return "No increment"

        elif curr_salary <= 10:
            if curr_grade == 'A':
                return curr_salary * 1.14
            elif curr_grade == 'B':
                return curr_salary * 1.10
            elif curr_grade == 'C':
                return curr_salary * 1.08
            elif curr_grade == 'D':
                return curr_salary * 1.06
            else:
                return "No increment"

        elif curr_salary <= 15:
            if curr_grade == 'A':
                return curr_salary * 1.08
            elif curr_grade == 'B':
                return curr_salary * 1.06
            elif curr_grade == 'C':
                return curr_salary * 1.04
            elif curr_grade == 'D':
                return curr_salary
            else:
                return "No increment"

        elif curr_salary <= 23:
            if curr_grade == 'A':
                return curr_salary * 1.07
            elif curr_grade == 'B':
                return curr_salary * 1.05
            elif curr_grade == 'C':
                return curr_salary * 1.04
            elif curr_grade == 'D':
                return curr_salary
            else:
                return "No increment"

        else:
            return "Salary out of range.."

    else:
        return "Invalid salary input"


# Taking input
curr_salary = input("Enter your salary in LPA: ")
curr_grade = input("Enter your grade A,B,C,D: ").upper()

cur_sal = salary_update(curr_salary, curr_grade)
print("Updated Salary:", cur_sal)
```

```python
from datetime import datetime

Interest_rates = {
    "7-45":{"Public_existing":5.75,"Public_revised":5.75,"Senior_existing":6.25,"Senior_revised":6.25},
    "46-179":{"Public_existing":6.25,"Public_revised":6.25,"Senior_existing":6.75,"Senior_revised":6.75},
    "180-210":{"Public_existing":6.35,"Public_revised":6.35,"Senior_existing":6.85,"Senior_revised":6.85},
    "211-less_than_1Year":{"Public_existing":6.40,"Public_revised":6.40,"Senior_existing":6.90,"Senior_revised":6.90},
    "1year-less_than_2_year":{"Public_existing":6.80,"Public_revised":7.00,"Senior_existing":7.30,"Senior_revised":7.50},
    "2years-less_than_3Year":{"Public_existing":6.80,"Public_revised":6.75,"Senior_existing":7.30,"Senior_revised":7.25},
    "3years-less_than_5years":{"Public_existing":6.80,"Public_revised":6.70,"Senior_existing":7.30,"Senior_revised":7.20},
    "5year-upto_10years":{"Public_existing":6.85,"Public_revised":6.60,"Senior_existing":7.35,"Senior_revised":7.10}
}

Time_brackets = {
    "7-45":(7,45),
    "46-179":(46,179),
    "180-210":(180,210),
    "211-less_than_1Year":(211,365),
    "1year-less_than_2_year":(365,730),
    "2years-less_than_3Year":(730,1095),
    "3years-less_than_5years":(1095,1826),
    "5year-upto_10years":(1826,3652)
}

def Your_tenure(days):
    for Tenors, (min_days, max_days) in Time_brackets.items():
        if min_days <= days <= max_days:
            return Tenors
    return "Invalid Duration"

start_date = input("Enter FD Start Date (YYYY-MM-DD): ")
end_date = input("Enter FD End Date (YYYY-MM-DD): ")
start_date = datetime.strptime(start_date, "%Y-%m-%d")
end_date = datetime.strptime(end_date, "%Y-%m-%d")

# difference
duration = (end_date - start_date).days
print("Total Days:", duration)
tenor = Your_tenure(duration)

if tenor != "Invalid Duration":
    rate = Interest_rates[tenor]['Public_existing']
    print("Tenor:", tenor)
    print("Interest Rate:", rate)
else:
    print("Invalid FD duration")
```

```python
#Q 50.	Write a code to generate factorial of the number  For example: factorial of 5 = 5! = 5*4*3*2*1
def fact(n):
    f = 1
    for i in range(1, n + 1):
        f = f * i
    print(f)

fact(int(input("Enter a number: ")))
```

```python
# Q52 Write a program to check frequency of each element in the list.
lst = [1, 2, 2, 3, 4, 1, 2]
d = {}
for i in lst:
    if i in d:
        d[i] = d[i] + 1
    else:
        d[i] = 1
print(d)
```
